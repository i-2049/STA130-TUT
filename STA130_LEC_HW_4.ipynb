**Session 1 Summary**: 
In this session, we discussed the **Standard Error of the Mean (SEM)** and how it is used to calculate a **95% confidence interval**. Here's a summary:

1. **SEM Definition**: SEM estimates how much the sample mean is likely to differ from the population mean. It's calculated by dividing the sample's standard deviation by the square root of the sample size.
   
2. **Confidence Interval Calculation**: A 95% confidence interval for the mean is calculated as:
   \[
   \text{CI} = \text{Sample Mean} \pm (Z \times SEM)
   \]
   where the Z-score for 95% confidence is approximately 1.96.

3. **Example**: We calculated SEM using a sample standard deviation of 10 and a sample size of 25. Then, using a sample mean of 50, we computed a 95% confidence interval as (46.08, 53.92).

This process helps estimate the range in which the true population mean likely falls, based on the sample data.

The "Pre-lecture" video (above) mentioned that bootstrap confidence intervals could apply to other statistics of the sample, such as the "median". 

Work with a ChatBot to create code to produce a 95% bootstrap confidence interval for a population mean based on a sample that you have and comment the code to demonstrate how the code can be changed to produce a 95% bootstrap confidence interval for different population parameter (other than the population mean, such as the population median).



```python
import numpy as np

# code creates a fuction to calculate the given confidence level (95%) of a sample bootstrapped a given number of times (10000)
def bootstrap_ci(sample, n_bootstrap=10000, confidence_level=0.95):
    # calculates bootstrapped means 
    bootstrap_means = np.array([np.mean(np.random.choice(sample, size=len(sample), replace=True)) for _ in range(n_bootstrap)])

    # upper and lower bounds calculated by taking the percentile of the b
    lower_bound = np.percentile(bootstrap_means, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_means, (1 + confidence_level) / 2 * 100)
    
    # function returns calculated the range of the confidence interval (upper and lower bounds) 
    return lower_bound, upper_bound

# sample generated by ChatGPT, this sample should be otherwise replaced by the desired sample 
sample = [12, 7, 10, 6, 9, 14, 8, 15] 

# running the sample data through the bootstrapping_ci function 
ci = bootstrap_ci(sample)

# output print confidence interval and additional text
print(f"95% Bootstrap Confidence Interval: {ci}")
```

# Question 6

1. The process of bootstrapping involves sampling from a dataset with replacement. This means each value you select from the original sample is automatically replenished, and each time you take a data value for the "new" sample (the bootstrapped sample), you are selecting it from the unaltered original data sample. This essentially simulates the sample, treating the original sample as if it were the whole population. When you have collected a new sample with the same number of values as the original sample, you calculate the statistics of that sample and record them in a histogram. This process is repeated, and eventually, there are enough sample sets taken from the original data to replicate the population as closely as possible. You can draw conclusions from the histogram created and infer average probabilities based on the information drawn from the data.

2. The main purpose of bootstrapping is to replicate the experiment as many times as possible and to conclude results from a variety of information based on each bootstrapping result. This approach saves money and time compared to actually redoing the experiment.

3. If I had a hypothesized guess of the average of a population and a sample size from that population, I would use bootstrapping to replicate the population. Running a large number of simulations by sampling from the original sample would, as accurately as possible, replicate the population. By observing the averages of each of these samples, I would be able to determine the confidence interval of the data. Then, by comparing my hypothesis with the confidence interval, I would be able to tell whether or not it is a plausible hypothesis.

```python
# CSV file
url = "https://raw.githubusercontent.com/i-2049/STA130-TUT/refs/heads/main/file.csv"
data = pd.read_csv(url)

# Calculate Change in Health Score
data['ScoreDifference'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Extract Change in Health Score for calculations
change_sample = data['ScoreDifference'].tolist()

def bootstrap_ci(sample, n_bootstrap=10000, confidence_level=0.95):
    bootstrap_means = np.array([np.mean(np.random.choice(sample, size=len(sample), replace=True)) for _ in range(n_bootstrap)])
    lower_bound = np.percentile(bootstrap_means, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_means, (1 + confidence_level) / 2 * 100)
    return lower_bound, upper_bound

# Calculate statistics for Change in Health Score
mean_change = np.mean(change_sample)
median_change = np.median(change_sample)
std_dev_change = np.std(change_sample)

# Run the sample data through the bootstrapping_ci function for Change in Health Score
ci_change = bootstrap_ci(change_sample)

# Output: print statistics and confidence intervals for Change in Health Score
print(f"Statistics for Change in Health Score:")
print(f"Mean: {mean_change:.2f}")
print(f"Median: {median_change:.2f}")
print(f"Standard Deviation: {std_dev_change:.2f}")
print(f"\n95% Bootstrap Confidence Interval for Change in Health Score: {ci_change}")
```
